```python
from PIL import Image
import os
import numpy as np
from keras.preprocessing.image import ImageDataGenerator
```

```python
def process_image(image_path):
    image = Image.open(image_path)
    # Resize
    resized_image = image.resize((28, 28))
    # Convert to grayscale
    bw_image = resized_image.convert("L")
    return bw_image
```

```python
def augment_images_60000(sample_images):
    # Data augmentation configuration
    datagen = ImageDataGenerator(
        zoom_range=0.2,
        horizontal_flip=True
    )

    augmented_images = []

    # Define the number of rotations and augmentations per rotation
    num_rotations = 4  # 0, 90, 180, 270 degrees
    augmentations_per_rotation = 150  # Adjust this as needed

    for rotation in range(num_rotations):
        for image in sample_images:
            rotated_image = image.rotate(rotation * 90)
            rotated_image_array = np.array(rotated_image).reshape((1, 28, 28, 1))

            # Producing augmented versions for each rotation
            for _ in range(augmentations_per_rotation):
                for batch in datagen.flow(rotated_image_array, batch_size=1):
                    augmented_image = Image.fromarray(batch[0].astype(np.uint8).reshape(28, 28))
                    augmented_images.append(augmented_image)
                    break

    return augmented_images
```

```python
def compute_label(image_array, whitespace_threshold=200):
    # Identify whitespace
    whitespace_mask = image_array > whitespace_threshold

    # Label distinct whitespace regions
    labeled_whitespace, num_features = label(whitespace_mask)

    # Compute size of each whitespace region
    whitespace_sizes = [np.sum(labeled_whitespace == i) for i in range(1, num_features + 1)]
    largest_whitespace = max(whitespace_sizes) if whitespace_sizes else 0

    return f"{num_features}_{largest_whitespace}"
```

```python
img_width, img_height =28,28
img_channel = 1#grayscale image
img_shape = (img_width, img_height, img_channel)
num_classes = 2010
z_dim = 100
```

```python
from keras.layers import UpSampling2D, Reshape, Activation, Conv2D, BatchNormalization, LeakyReLU, Input, Flatten, multiply,ReLU
from keras.layers import Dense, Embedding
from keras.models import Sequential, Model
from keras.constraints import Constraint
import keras.backend as K
from keras.layers import Conv2DTranspose
from keras.layers import RepeatVector


# Define a custom max-norm constraint
# class MaxNorm(Constraint):
#     def __init__(self, max_value=1, axis=0):
#         self.max_value = max_value
#         self.axis = axis

#     def __call__(self, w):
#         norms = K.sqrt(K.sum(K.square(w), axis=self.axis, keepdims=True))
#         desired = K.clip(norms, 0, self.max_value)
#         return w * (desired / (K.epsilon() + norms))


def build_generator():
    model = Sequential()
    model.add(Dense(128*7*7, activation = 'relu', input_shape = (z_dim, )))
    model.add(Reshape((7,7,128)))
    model.add(UpSampling2D())

    # , kernel_constraint=MaxNorm(1.0)
    model.add(Conv2D(128, kernel_size = 3, strides = 2, padding = 'same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha = 0.02))
    model.add(UpSampling2D())#

    model.add(Conv2D(128, kernel_size = 3, strides = 1, padding = 'same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha = 0.02))
    model.add(UpSampling2D())

    model.add(Conv2D(1, kernel_size = 3 , strides = 1, padding='same'))
    model.add(Activation('tanh'))

    z = Input(shape= (z_dim,))
    label = Input(shape=(1,), dtype = 'int32')

    label_embedding = Embedding(num_classes, z_dim, input_length = 1)(label)
    label_embedding = Flatten()(label_embedding)
    # label_embedding = RepeatVector(1)(label_embedding)  # Repeat for batch size
    joined = multiply([z, label_embedding])

    img = model(joined)
    return Model([z, label], img)

generator = build_generator()
generator.summary()
```

```python
from keras.layers import Dropout, Concatenate, MaxPool2D
import numpy as np
from keras.regularizers import l2


def build_discriminator():
    model = Sequential()
    model.add(Conv2D(32, kernel_size=3, strides=2, input_shape=(28, 28, 2), padding='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha=0.02))

    model.add(Conv2D(64, kernel_size=3, strides=2, padding='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha=0.02))

    model.add(Conv2D(128, kernel_size=3, strides=2, padding='same'))
    model.add(BatchNormalization())
    model.add(LeakyReLU(alpha=0.02))
    model.add(Dropout(0.2))

    model.add(Flatten())
    model.add(Dense(1, activation='sigmoid', kernel_regularizer=l2(0.01)))  # L2 regularization

    img = Input(shape=img_shape)
    label = Input(shape=(1,), dtype='int32')

    # Use an appropriate output_dim to match img_shape
    label_embedding = Embedding(input_dim=num_classes, output_dim=np.prod(img_shape))(label)

    label_embedding = Flatten()(label_embedding)
    label_embedding = Reshape(img_shape)(label_embedding)

    concat = Concatenate(axis=-1)([img, label_embedding])
    prediction = model(concat)
    return Model([img, label], prediction)

discriminator = build_discriminator()
discriminator.summary()
```

```python
from keras.optimizers import Adam

discriminator.compile(loss = 'binary_crossentropy', optimizer = Adam(0.0001, 0.5), metrics = ['accuracy'])

z = Input(shape=(z_dim,))
label = Input(shape= (1,))
img = generator([z,label])

discriminator.trainable = False
prediction = discriminator([img, label])

cgan = Model([z, label], prediction)
cgan.compile(loss= 'binary_crossentropy', optimizer = Adam(0.0001, 0.5))
```

```python
def train(epochs, batch_size, save_interval,X_train,y_train):

    # X_train = (X_train - 127.5) / 127.5
    X_train = np.expand_dims(X_train, axis=3)

    real = np.ones(shape=(batch_size, 1))#array of ones, representing real images
    fake = np.zeros(shape=(batch_size, 1))

    for iteration in range(epochs):

        idx = np.random.randint(0, X_train.shape[0], batch_size)
        imgs, labels = X_train[idx], y_train[idx]

        z = np.random.normal(0, 1, size=(batch_size, z_dim))
        gen_imgs = generator.predict([z, labels])

        d_loss_real = discriminator.train_on_batch([imgs, labels], real)
        d_loss_fake = discriminator.train_on_batch([gen_imgs, labels], fake)
        d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)#balance the learning between the generator and discriminator.

        z = np.random.normal(0, 1, size=(batch_size, z_dim))
        labels = np.random.randint(0, num_classes, batch_size).reshape(-1, 1)

        g_loss = cgan.train_on_batch([z, labels], real)

        if iteration % save_interval == 0:
            print('{} [D loss: {}, accuracy: {:.2f}] [G loss: {}]'.format(iteration, d_loss[0], 100 * d_loss[1], g_loss))
            save_image(iteration)
```
